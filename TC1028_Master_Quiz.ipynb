**Problema**

Desarrolla un programa en Python que realice los siguientes puntos:

Muestre al usuario las siguientes 3 opciones:

Opción 1 - Cuadrado


Opción 2 - Círculo

Opción 3 - Rectángulo

1. Pida al usuario seleccionar una opción y con base en su elección, preguntar la información necesaria para calcular el área de esa figura. (15 puntos)
2. Valida si el usuario capturó lo necesario para calcular el área de la figura, si es así, calcula e imprime el área de la figura dada, si no  imprime la palabra "Figura invalida". (20 puntos)
3. Pregunta si desea volver a calcular el área de otra figura para desplegar de nuevo las opciones y si no salir del programa. (20 puntos)
4. Al finalizar, muestra el promedio del área de las figuras calculadas así como un mensaje de despedida. (15 puntos)

Recuerda que:

La base y altura del cuadrado son números enteros positivos iguales entre si, mayores a cero

La base y altura del rectángulo son números enteros positivos diferentes entre si, mayores a cero

El radio del círculo es un número entero positivo mayor a cero

Área del cuadrado y rectángulo = base * altura

Área del círculo = PIr2

Considera el valor de PI = 3.14

```
suma_areas = 0 # la usaremos para sumar las áreas de las figuras para luego calcular el promedio, inicia en cero
figuras = 0 # será el contador de figuras para luego calcular el promedio, inicia en cero
area = 0 # para este caso es necesario iniciar en cero, para borrar valores de corridas anteriores

while True: # creamos un ciclo while infinito con True, para que pregunte indefinidamente, es decir, correrá hasta que le digamos break
  print("Elige el número de la figura para calcular su área:","1 - Cuadrado","2 - Círculo","3 - Rectángulo", sep="\n") # imprimimos el menú, sep="\n" es para que cambie de línea
  num = int(input()) # preguntamos por el número de figura

  if num == 1: # validamos si el número de la figura es 1 entonces solicitamos los datos para calcular el área de un cuadrado
    print("Ingresa el tamaño de un lado")
    lado = int(input()) # preguntamos el lado

    if lado > 0: # validamos que sea número mayor que cero
      area = lado * lado # calculamos área
      print("El área del cuadrado es", area)
      figuras = figuras + 1 # le agregamos 1 al contador de figuras cada vez que calculamos el área
    else: # si no es mayor que cero entonces es figura inválida
      print("Figura inválida")
      continue # regresa al inicio del while y repite el proceso

  elif num == 2: # validamos si el número de la figura es 2 entonces solicitamos los datos para calcular el área de un círculo
    print("Ingresa el tamaño del radio")
    radio  = int(input()) # preguntamos el radio
    
    if radio > 0: # validamos que sea número mayor que cero
      area = 3.14 * radio ** 2 # calculamos área
      print("El área del círculo es", area)
      figuras = figuras + 1 # le agregamos 1 al contador de figuras cada vez que calculamos el área
    else: # si no es mayor que cero entonces es figura inválida
      print("Figura inválida")
      continue # regresa al inicio del while y repite el proceso
    
  elif num == 3: # validamos si el número de la figura es 3 entonces solicitamos los datos para calcular el área de un rectágulo
    print("Ingresa el tamaño de la base")
    base = int(input()) # preguntamos la base
    print("Ingresa el tamaño de la altura")
    altura = int(input()) # preguntamos la altura

    if base > 0 and altura > 0 and base != altura: # validamos que sean número mayor que cero, y que la base y la altura sean diferentes para considerarlo rectángulo
      area = base * altura # calculamos área
      print("El area del rectándulo es", area)
      figuras = figuras + 1 # le agregamos 1 al contador de figuras cada vez que calculamos el área
    else: # si no es mayor que cero y base y altura son iguales entonces es figura inválida
      print("Figura inválida")
      continue # regresa al inicio del while y repite el proceso
  
  else: # si no elige una de las 3 opciones del menú se lo indicamos
    print("Opción incorrecta... intenta ingresar uno de los números del menú de figuras")

  suma_areas = suma_areas + area # agregamos el area calculada en el ciclo actual al total de áreas

  print("Desea calcular otra área Si/No")
  continuar = input() # preguntamos si desea continuar
  
  if continuar == "si":
    continue # si dice que "si" regresa al inicio del while y repite el proceso
  else: # si dice otra cosa
    if figuras == 0: # si el programa no ha calculado áreas el contador de figuras es cero, por lo tanto no podemos desplegar promedio
      print("No se calcularon áreas de figuras para promediar. Hasta luego")
    else: # si el contador de figuras no está en cero, entonces calculamos el promedio
      print("El promedio del área de las figuras es:", suma_areas/figuras,"Hasta luego", sep="\n")
    break # salimos del ciclo while y termina el programa
```
